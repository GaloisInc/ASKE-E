```
sir = loadESL("sir.easel")
sir_small = loadESL("sir-small-beta.easel")
sir_ensemble = ensemble([[sir, 0.8], [sir_small, 0.2]])
```

```
describeModel(sir_ensemble)
```

```
points = simulate(sir_ensemble at [0..200 by 1])

S_large = series(value(points.S_1), "S_large", {{ style = "lines" }})
I_large = series(value(points.I_1), "I_large", {{ style = "lines" }})
R_large = series(value(points.R_1), "R_large", {{ style = "lines" }})

S_small = series(value(points.S_2), "S_small", {{ style = "lines" }})
I_small = series(value(points.I_2), "I_small", {{ style = "lines" }})
R_small = series(value(points.R_2), "R_small", {{ style = "lines" }})

S_average = series(value(points.S), "S_average", {{ style = "lines" }})
I_average = series(value(points.I), "I_average", {{ style = "lines" }})
R_average = series(value(points.R), "R_average", {{ style = "lines" }})
```

```
plot("", [S_large, I_large, R_large], time(points), "time")
```

```
plot("", [S_small, I_small, R_small], time(points), "time")
```
