```
sir = loadESL("sir.easel")
sirsd = loadESL("sirs.easel")
joined = join(["SIR_", sir], ["SIRSD_", sirsd], [["S", "S"]])
```

```
describeModel(joined)
```

```
modelGraph(sir)
```

```
modelGraph(sirsd)
```

```
modelGraph(joined)
```

```
sirPoints = simulate(sir at [0..100 by 1])

S = series(value(sirPoints.S), "S", {{ style = "lines" }})
I = series(value(sirPoints.I), "I", {{ style = "lines" }})
R = series(value(sirPoints.R), "R", {{ style = "lines" }})

plot("", [S, I, R], time(sirPoints), "time")
```

```
sirsdPoints = simulate(sirsd at [0..100 by 1])

S = series(value(sirsdPoints.S), "S", {{ style = "lines" }})
I = series(value(sirsdPoints.I), "I", {{ style = "lines" }})
R = series(value(sirsdPoints.R), "R", {{ style = "lines" }})
D = series(value(sirsdPoints.D), "D", {{ style = "lines" }})

plot("", [S, I, R, D], time(sirsdPoints), "time")
```

```
points = simulate(joined at [0..100 by 1])

S = series(value(points.S), "S", {{ style = "lines" }})
SIR_I = series(value(points.SIR_I), "SIR_I", {{ style = "lines" }})
SIR_R = series(value(points.SIR_R), "SIR_R", {{ style = "lines" }})

SIRSD_I = series(value(points.SIRSD_I), "SIRSD_I", {{ style = "lines" }})
SIRSD_R = series(value(points.SIRSD_R), "SIRSD_R", {{ style = "lines" }})
SIRSD_D = series(value(points.SIRSD_D), "SIRSD_D", {{ style = "lines" }})

plot("", [S, SIR_I, SIR_R, SIRSD_I, SIRSD_R, SIRSD_D], time(points), "time")
```
